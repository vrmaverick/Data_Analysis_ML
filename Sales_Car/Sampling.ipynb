## Handling Missing Values

```python
missingData_csv = pd.read_csv("./data/car_missing_data.csv")
missingData_csv
```

```python
missingData_csv["Odometer"].fillna(missingData_csv["Odometer"].mean(),inplace=True) #inplace is used for saving data to missingData_csv
missingData_csv # NaN replaced by mean for one column
```

```python
#Dropping other NaN columns
new_data = missingData_csv.dropna(inplace=False) #defualt inplace=False prevents overwriting the changes on main missingData_csv
new_data #new data holds all filled values
```

# Adding new Data 

```python
Seats = pd.Series([8,5,2,5])
new_data["Seats"]=Seats
new_data
```

So if the series has less values it still works and other entries have NaN which can be resolved


```python
new_data["Seats"].fillna(5,inplace=True)
```

```python
new_data
```

Using lists to add data of n length where n is the length of the the dataframe

```python
HorsePower = [223,556,789,456,789,323]
new_data["Horse_power"] = HorsePower # dataframe size and list size should be equal
new_data
```

```python
new_data["Wheels"] = 4
new_data["Emissions per liter(kg)"]=1.25
Economy = [12,56,67,86,33,23]
new_data["Per liter Economy"] = Economy
```

```python
new_data
```

```python
new_data['total emission'] = (new_data['Odometer']/new_data['Per liter Economy'])*new_data['Emissions per liter(kg)']
new_data 
```

```python
new_data["total emission"].plot()
```

```python
new_data = new_data.drop('Doors',axis=1) #delete column that is not required
new_data
```

# Sampling the dataset


```python
sampled = new_data.sample(frac=0.8) #80% data for  sample with randomized order
```

```python
sampled
```

```python
sampled = sampled.reset_index(drop=True)
sampled
```

Applying Functions to diffrent columns


```python
sampled.rename(columns={'Odometer':'Miles'},inplace= True)
```

```python
sampled['miles'] = sampled['miles'].apply(lambda x:x/1.6) #converting Km to miles
```
