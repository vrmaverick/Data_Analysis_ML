```python
x =np.linspace(0,10,100) #return a evenly spaced numbers over specified interval
```

```python
x[:10]
```

```python
fig,ax = plt.subplots()
ax.plot(x,x**3)
```

```python
#Scatter Plot
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
#Tanh wave
fig,ax = plt.subplots()
ax.scatter(x,np.tanh(x))
```

```python
# Vertical Bar Plot
price = {'Tea':90,'Coffee':56,'Sugar':50} #price list
fig,ax = plt.subplots()
ax.bar(price.keys(),price.values());
ax.set(title="Cost of Cafe's");
```

```python
#Horizontal Bar Plot :Takes attributes as Lists
fig,ax = plt.subplots()
ax.barh(list(price.keys()),list(price.values()));
ax.set(title="Cost of Cafe's");
```

```python
#Histogram
normal_sample = np.random.randn(10000) #returns 1000 elements from normal distribution
normal_sample[:10]
```

```python
fig,ax = plt.subplots()
ax.hist(normal_sample) #normal continuous distribution
```

## Subplots
Option 1 : Either by defining axes objects <br>
Option 2 : Using lsting of sxes same but syntax is diffrent both examples are tested below

```python
fig,((axes1,ax2),(ax3,ax4)) = plt.subplots(ncols=2,nrows=2,figsize=(10,5))
axes1.scatter(price.keys(),price.values())
ax2.plot(price.keys(),price.values())
ax3.hist(price.values())
ax4.bar(price.keys(),price.values());

```

```python
fig,ax = plt.subplots(ncols=2,nrows=2,figsize=(10,5))
ax[0,0].scatter(price.keys(),price.values())
ax[0,1].plot(price.keys(),price.values())
ax[1,0].hist(price.values())
ax[1,1].bar(price.keys(),price.values());
```
